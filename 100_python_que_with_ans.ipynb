```python
#47. Write a program to check whether a number is an Armstrong number

n=int(input('Enter a number:'))
n=abs(n)
n=str(n)
k=len(n)
arm=0
for i in n:
  arm=arm+int(i)**k
if arm==int(n):
  print(n,'is armstrong number')
else:
  print(n,'is not armstrong number')
```

```python
#48. Write a program to display all Armstrong numbers from 1 to n.

n=int(input('Enter a number:'))
for num in range(1,n+1):
  num=str(num)
  k=len(num)
  arm=0
  for i in num:
    arm=arm+int(i)**k
  if arm==int(num):
    print(num,end=" ")
```

```python
#49. Write a program to check whether a number is a perfect number.

num = int(input('Enter the number='))

if num<=1:
  print('It is -ve number')
else:
  divisor_sum=sum(i for i in range(1,(num)) if num%i==0)
  if divisor_sum==num:

    print(num,'is perfect number')
  else:
    print(num,'is not perfect number')
```
