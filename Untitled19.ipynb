```python
items = {'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill = []
def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

# calling of purchase
total = purchase('wheat','milk')
print('Item amounts:',total)

# calling of check
ans = check(total)
print('Final bill amoutnt is Rs:',ans)
```

```python
def calculate_per(marks_obtained, marks):
  return (marks_obtained / marks) * 100
#print(calculate_per(444,500))

def calculate_grade(per):
  if per >= 75:
    return 'Distinction'
  elif per >= 65:
    return 'First class'
  elif per>=55:
    return 'Second class'
  elif per >= 45:
      return 'Pass class'
  else:
    return 'Fail'

per = calculate_per(444,500)
print('Total percentage:',per)
grade = calculate_grade(per)
print('Your grade:',grade)

```
